### Dataset 2

```python
df1 = pd.read_csv('datasets/sentences/2_darija_ar_sentences.csv')
df2 = pd.read_csv('datasets/sentences/MAC corpus.csv')
```

```python
print(df2.shape)
print(df2.info())
df2.head()
```

```python
df2[df2['class'] == 'dialectal'].value_counts().sum()
```

```python
df2_selected = df2[df2['class'] == 'dialectal'][['tweets', 'type']]
df2_selected.rename(columns={
    'tweets': 'darija_ar',
    'type': 'Sentiment'
}, inplace=True)
```

```python
import pandas as pd

def append_and_save(df1, df2, save_path):
    """
    Appends two dataframes and saves the resulting dataframe to a CSV file.

    Parameters:
    - df1 (pd.DataFrame): The first dataframe.
    - df2 (pd.DataFrame): The second dataframe to append to the first.
    - save_path (str): Path to save the appended dataframe as a CSV file.

    Returns:
    - pd.DataFrame: The appended dataframe.
    """
    # Append df2 to df1
    appended_df = pd.concat([df1, df2], axis=0, ignore_index=True)

    # Save the appended dataframe to the specified path
    appended_df.to_csv(save_path, index=False)

    # Print the shape and first few rows of the appended dataframe
    print(f"Shape of appended dataframe: {appended_df.shape}")
    print(appended_df.head())

    return appended_df
```

```python
append_and_save(df1, df2_selected, 'datasets/sentences/3_darija_ar_sentences.csv')
```
