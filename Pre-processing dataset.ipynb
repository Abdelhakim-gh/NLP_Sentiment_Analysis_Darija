PyArabic fournit des outils utiles pour le traitement de texte en arabe, incluant des fonctions pour nettoyer et normaliser les textes

*   utilisée pour supprimer le caractère de tatweel (التطويل), aussi appelé kashida (ـــــ)
*   Cette fonction supprime les diacritiques (tashkeel - تشكيل) du texte arabe.



```python
import re
import emoji
def contains_emoji(s):
    if isinstance(s, str):  # Check if the input is a string
        return bool(re.search(r'[^\w\s,]', s))
    return False  # Return False for non-string inputs

# Filter sentences containing emojis
emoji_sentences = df1[df1['sentence'].apply(contains_emoji)]

# Display sentences with emojis
print("Sentences containing emojis:")
print(emoji_sentences)
```

```python
import pandas as pd
import re
import emoji
from pyarabic.araby import strip_tatweel, strip_tashkeel

# Function to remove emojis
def remove_emojis(text):
    if not isinstance(text, str):  # Check if text is a string
        return ""
    # Remove the ♥ emoji and other emojis using the emoji library
    text = text.replace('♥', '')  # Specifically remove the ♥ emoji
    return emoji.replace_emoji(text, replace='')  # Remove all other emojis

# Function to clean the text
def clean_text(text):
    if not isinstance(text, str):  # Check if text is a string
        return ""
    text = re.sub(r'http\S+', '', text)  # Remove URLs
    text = re.sub(r'@\w+', '', text)    # Remove mentions (e.g., @username)
    text = re.sub(r'[^\w\s\u0600-\u06FF]', '', text)  # Remove punctuation
    return text.strip()

# Function to remove repeated characters
def remove_repeated_chars(text):
    if not isinstance(text, str):  # Check if text is a string
        return ""
    return re.sub(r'(.)\1+', r'\1', text)

# Preprocessing: Ensure all values in the 'sentence' column are strings
df1['sentence'] = df1['sentence'].fillna("").astype(str)

# Apply the cleaning functions to the 'sentence' column
df1['cleaned_text'] = df1['sentence'].apply(remove_emojis)        # Remove emojis
df1['cleaned_text'] = df1['cleaned_text'].apply(clean_text)       # Clean the text
df1['cleaned_text'] = df1['cleaned_text'].apply(remove_repeated_chars)  # Remove repeated characters

# Apply tatweel and tashkeel removal
df1['cleaned_text'] = df1['cleaned_text'].apply(strip_tatweel)
df1['cleaned_text'] = df1['cleaned_text'].apply(strip_tashkeel)

# Display the first few rows of the cleaned DataFrame
print(df1.head())
```

```python
def contains_emoji(text):
    if not isinstance(text, str):  # Ensure the input is a string
        return False
    for char in text:
        if emoji.is_emoji(char):
            return True
    return False

# Create a cleaned_text column (if not already done)
if 'cleaned_text' not in df1.columns:
    df1['cleaned_text'] = df1['sentence']  # Example: Copy from sentence for testing

# Apply the emoji detection function
df1['contains_emoji'] = df1['cleaned_text'].apply(contains_emoji)

# Filter rows where emojis are found in cleaned_text
emoji_rows = df1[df1['contains_emoji']]

# Display rows with emojis
print(emoji_rows)
```
